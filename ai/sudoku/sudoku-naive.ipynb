```python
class SudokuCell:
    def __init__(self, pos, row, column, square):
        '''
        pos => int
        row => list
        column => list
        square => list
        '''
        self.pos = pos
        self.row = row
        self.column = column
        self.square = square
        listy = []
        for i in row:
            listy.append(i)
        for i in column:
            listy.append(i)
        for i in square:
            listy.append(i)
        self.neighborhood = set(listy)
        self.info = {pos: list(self.neighborhood)}
    def __str__(self):
        return str(self.pos)
```
